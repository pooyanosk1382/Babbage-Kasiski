# What is our goal?
Here we want to decrypt the Vigenere cipher.
For this aim we should use the Babbage-Kasiski test.

# What is Babbage-Kasiski test?
The Kasiski test was described by Friedrich Kasiski in 1863; however, it was apparently discovered earlier, around 1854, by Charles Babbage.
The Kasiski examination involves looking for strings of characters that are repeated in the ciphertext. The strings should be three characters long or more for the examination to be successful. Then, the distances between consecutive occurrences of the strings are likely to be multiples of the length of the keyword.

Here we start.
At first find the probability of each letter.

```python
import  PyPDF2
import matplotlib.pyplot as plt
import numpy as np
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
readerCLRS = PyPDF2.PdfReader('CLRS.pdf')
alphabetValuesOfCLRS = [0] * 26
probability = [0] * 26
indOfAlphabet = [0] * 26
```

```python
for page in range(1313):
    st = readerCLRS.pages[page].extract_text()
    for i in st:
        if i == 'a' or i == 'A':
            alphabetValuesOfCLRS[0] += 1
        elif i == 'b' or i == 'B':
            alphabetValuesOfCLRS[1] += 1
        elif i == 'c' or i == 'C':
            alphabetValuesOfCLRS[2] += 1
        elif i == 'd' or i == 'D':
            alphabetValuesOfCLRS[3] += 1
        elif i == 'e' or i == 'E':
            alphabetValuesOfCLRS[4] += 1
        elif i == 'f' or i == 'F':
            alphabetValuesOfCLRS[5] += 1
        elif i == 'g' or i == 'G':
            alphabetValuesOfCLRS[6] += 1
        elif i == 'h' or i == 'H':
            alphabetValuesOfCLRS[7] += 1
        elif i == 'i' or i == 'I':
            alphabetValuesOfCLRS[8] += 1
        elif i == 'j' or i == 'J':
            alphabetValuesOfCLRS[9] += 1
        elif i == 'k' or i == 'K':
            alphabetValuesOfCLRS[10] += 1
        elif i == 'l' or i == 'L':
            alphabetValuesOfCLRS[11] += 1
        elif i == 'm' or i == 'M':
            alphabetValuesOfCLRS[12] += 1
        elif i == 'n' or i == 'N':
            alphabetValuesOfCLRS[13] += 1
        elif i == 'o' or i == 'O':
            alphabetValuesOfCLRS[14] += 1
        elif i == 'p' or i == 'P':
            alphabetValuesOfCLRS[15] += 1
        elif i == 'q' or i == 'Q':
            alphabetValuesOfCLRS[16] += 1
        elif i == 'r' or i == 'R':
            alphabetValuesOfCLRS[17] += 1
        elif i == 's' or i == 'S':
            alphabetValuesOfCLRS[18] += 1
        elif i == 't' or i == 'T':
            alphabetValuesOfCLRS[19] += 1
        elif i == 'u' or i == 'U':
            alphabetValuesOfCLRS[20] += 1
        elif i == 'v' or i == 'V':
            alphabetValuesOfCLRS[21] += 1
        elif i == 'w' or i == 'W':
            alphabetValuesOfCLRS[22] += 1
        elif i == 'x' or i == 'X':
            alphabetValuesOfCLRS[23] += 1
        elif i == 'y' or i == 'Y':
            alphabetValuesOfCLRS[24] += 1
        elif i == 'z' or i == 'Z':
            alphabetValuesOfCLRS[25] += 1
plt.plot(alphabet, alphabetValuesOfCLRS)
plt.title('number of letters in book of CLRS')
```

```python
allVal = 0
for i in range(26):
    allVal += alphabetValuesOfCLRS[i]
allVal
```

```python
for i in range(26):
    probability[i] = alphabetValuesOfCLRS[i]/allVal
probability
```

```python
plt.plot(alphabet, probability)
```
